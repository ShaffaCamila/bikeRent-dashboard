### Cleaning Data

#### Membersihkan tabel `day_df`

Mengkonversi nilai representasi angka menjadi kategori

```python
day_df['season'] = day_df['season'].map({
    1: 'spring',
    2: 'summer',
    3: 'fall',
    4: 'winter'    
})


day_df['yr'] = day_df['yr'].map({
    0: 2011,
    1: 2012    
})

day_df['holiday'] = day_df['holiday'].map({
    0: 'no',
    1: 'yes'    
})


day_df['workingday'] = day_df['workingday'].map({
    0: 'no',
    1: 'yes'    
})

day_df['weathersit'] = day_df['weathersit'].map({
    1: 'Clear',
    2: 'Mist',
    3: 'Light Rain/Snow',
    4: 'Heavy Rain/Snow'    
})
```

Mengubah data type

```python
#Mengganti tipe data menjadi kategori
change_columns = ['season', 'holiday', 'workingday', 'weathersit']

for column in change_columns:
    day_df[column] = day_df[column].astype('category')
```

Rename

```python
#Mengganti nama kolom pada tabel day_df
day_df = day_df.rename(columns={'instant': 'record_index', 'dteday': 'date', 'yr': 'year', 'mnth': 'month', 'workingday': 'working_day', 'weathersit': 'weather', 'temp': 'temperature', 'atemp': 'feels_like_temp', 'hum': 'humidity', 'windspeed': 'wind_speed', 'casual': 'casual_users', 'registered': 'registered_users', 'cnt': 'total_count'})
```

```python
day_df.info()
```

#### Membersihkan tabel `hour_df`

```python
hour_df['season'] = hour_df['season'].map({
    1: 'spring',
    2: 'summer',
    3: 'fall',
    4: 'winter'    
})

hour_df['yr'] = hour_df['yr'].map({
    0: 2011,
    1: 2012    
})

hour_df['holiday'] = hour_df['holiday'].map({
    0: 'no',
    1: 'yes'    
})

hour_df['workingday'] = hour_df['workingday'].map({
    0: 'no',
    1: 'yes'    
})

hour_df['weathersit'] = hour_df['weathersit'].map({
    1: 'Clear',
    2: 'Mist',
    3: 'Light Rain/Snow',
    4: 'Heavy Rain/Snow'    
})
```

Mengubah data type

```python
#Mengganti tipe data menjadi kategori
change_columns = ['season', 'holiday', 'workingday', 'weathersit']

for column in change_columns:
    hour_df[column] = hour_df[column].astype('category')
```

Rename

## Exploratory Data Analysis (EDA)

### Explore `day_df`

```python
day_df.describe(include="all")
```

```python
yearly_trend = day_df.groupby('year', observed=False).agg({
    'total_count': ['max', 'min', 'sum', 'mean']
})
yearly_trend.head()
```

```python
monthly_progress = day_df.groupby(['month', 'year'], observed=False).agg({
    'total_count': 'sum'
}).reset_index()

monthly_progress['month/year'] = monthly_progress['month'].astype(str) + '/' + monthly_progress['year'].astype(str)

monthly_progress = monthly_progress.sort_values(by=['year', 'month']).reset_index(drop=True)

monthly_progress
```

```python
daily_trend = day_df.groupby('weekday', observed=False).agg({
    'total_count': ['max', 'min', 'sum', 'mean']
}).reset_index()
daily_trend['weekday'] = daily_trend['weekday'].map({
    0: 'Sunday',
    1: 'Monday',
    2: 'Tuesday',
    3: 'Wednesday',
    4: 'Thursday',
    5: 'Friday',
    6: 'Saturday'
})
daily_trend
```

```python
season_order = ['spring', 'summer', 'fall', 'winter']
day_df['season'] = pd.Categorical(day_df['season'], categories=season_order, ordered=True)

season_trend = day_df.groupby('season', observed=False).agg({
    'total_count': ['max', 'min', 'sum', 'mean']
})
season_trend
```

```python
weather_trend = day_df.groupby('weather', observed=False).agg({
    'total_count': ['max', 'min', 'sum', 'mean']
})
weather_trend
```

```python
month_trend = day_df.groupby('month', observed=False).agg({
    'total_count': ['max', 'min', 'sum', 'mean']
}).reset_index()
month_trend['month'] = month_trend['month'].map({
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'  
})
month_trend
```

```python
user_comparison = day_df.agg({
    'registered_users': ['sum', 'mean'],
    'casual_users': ['sum', 'mean']
}).map(lambda x: round(x, 2))
user_comparison
```

```python
workingday_trend = day_df.groupby('working_day', observed=False).agg({
    'total_count': ['sum', 'mean']
})
workingday_trend
```

### Explore `hour_df`

```python
hour_df.describe(include="all")
```

```python
hourly_trend  = hour_df.groupby('hour', observed=False).agg({
    'total_count': ['max', 'min', 'sum', 'mean']
}).reset_index()
hourly_trend
```

**Insight:**
- Persebaran data terjadi pada tahun 2011 hingga 2012
- Tercatat penyewaan berdasarkan tahun, bulan, hari, jam, musim, katgori pengguna, dll

```python
day_df["date"] = pd.to_datetime(day_df["date"])

rfm_df = day_df.groupby(by="month", as_index=False).agg({
    "date": "max",  
    "record_index": "nunique", 
    "total_count": "sum" 
})
rfm_df.columns = ["month", "max_order_timestamp", "frequency", "monetary"]

rfm_df["max_order_timestamp"] = pd.to_datetime(rfm_df["max_order_timestamp"]).dt.date

recent_date = day_df["date"].max().date()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("max_order_timestamp", axis=1, inplace=True)

rfm_df = rfm_df[["month", "recency", "frequency", "monetary"]]
print(rfm_df)
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_progress['month/year'],
    monthly_progress['total_count'],
    marker='o',
    linewidth=2,
    label='Monthly Progress from 2011 to 2012'
)
plt.title('Monthly Progress of Bike Renting (2011-2012)')
plt.xlabel('month/year')
plt.ylabel('Total Bike Rented')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

**Insight:**
- Terjadi kenaikan pesat pada bulan 2/2012 - 3/2012
- Terjadi penurunan drastis pada bulan 9/2012 - 10/2012
- Pada tahun 2012 mengalami peningkatan dibanding tahun 2011

### Pertanyaan 2:

```python
max_value = month_trend[('total_count', 'mean')].max()

colors = ['#A7C7E7' if v < max_value else '#004C99' for v in month_trend[('total_count', 'mean')]]

plt.figure(figsize=(10, 5))
sns.barplot(
    x='month',
    y=('total_count', 'mean'),
    data=month_trend,
    palette=colors  
)
plt.title('Total Bike Rented by Month')
plt.xlabel('Month')
plt.ylabel('Total Bike Rented')
plt.xticks(rotation=45)
plt.grid(False)
plt.show()
```

```python
max_value = daily_trend[('total_count', 'mean')].max()

colors = ['#A7C7E7' if v < max_value else '#004C99' for v in daily_trend[('total_count', 'mean')]]

plt.figure(figsize=(10, 5))
sns.barplot(
    x='weekday',
    y=('total_count', 'mean'),
    data=daily_trend,
    palette=colors  
)
plt.title('Total Bike Rented by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Total Bike Rented')
plt.grid(False)
plt.show()
```

```python
max_value = season_trend[('total_count', 'mean')].max()

colors = ['#A7C7E7' if v < max_value else '#004C99' for v in season_trend[('total_count', 'mean')]]

plt.figure(figsize=(10, 5))
sns.barplot(
    x='season',
    y=('total_count', 'mean'),
    data=season_trend,
    palette=colors 
)
plt.title('Total Bike Rented by Season of the Year')
plt.xlabel('Season of the Year')
plt.ylabel('Total Bike Rented')
plt.grid(False)
plt.show()
```

**Insight:**
- Pada awal bulan hingga pertengahan tahun menggalami penurunan, namun menggalami penurunan di beberapa bulan sebelum akhir tahun.
- Hari Jumat menjadi hari terbanyak terjadi penyewaan sepeda. Setelah melewati hari Selasa pengguna menjadi lebih banyak walaupun tidak terdapat lonjakan yang signifikan.
- Musim yang paling digemari para pengguna adalah Fall.

### Pertanyaan 3:

```python
max_value = hourly_trend[('total_count', 'sum')].max()

colors = ['#A7C7E7' if v < max_value else '#004C99' for v in hourly_trend[('total_count', 'sum')]]

hourly_trend['hour_formatted'] = hourly_trend['hour'].astype(str).str.zfill(2) + ".00"

plt.figure(figsize=(10, 5))
sns.barplot(
    x='hour_formatted', 
    y=('total_count', 'sum'),
    data=hourly_trend,
    palette=colors
)
plt.title('Total Bike Rented by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Bike Rented')
plt.xticks(rotation=90)
plt.grid(False)
plt.show()
```

**Insight:**
- Pada pukul 06.00 terjadi traffic yang cukup ramai pada penyewaan.
- Mulai pukul 09.00 hingga 16.00 dapat dikatakan stabil walaupun terdapat turun-naik namun tidak secara signifikan.

### Pertanyaan 4:

```python
values = [user_comparison['registered_users']['mean'], user_comparison['casual_users']['mean']]
labels = ['Registered', 'Casual']
colors = ['#A7C7E7', '#004C99']  

plt.figure(figsize=(8, 8))  
plt.pie(
    x=values,
    labels=labels,
    autopct='%1.1f%%',
    colors=colors,  
    startangle=90,  
    explode=[0.05, 0.05],  
    shadow=True, 
    textprops={'fontsize': 14}
)
plt.title('Total Bike Rented by User Type', fontsize=16)
plt.show()
```

**Insight:**
- Jenis pengguna yang paling banyak atau unggul adalah pengguna yang terdaftar (registered)

### Pertanyaan 5:

```python
values = [workingday_trend['total_count']['mean'][0], workingday_trend['total_count']['mean'][1]]
labels = ['Holiday', 'Working Day']
colors = ['#A7C7E7', '#004C99']

plt.figure(figsize=(8, 8))
plt.pie(
    x=values,
    labels=labels,
    autopct='%1.1f%%',
    colors=colors, 
    startangle=90,  
    explode=[0.05, 0.05],  
    shadow=True,  
    textprops={'fontsize': 14} 
)
plt.title('Total Bike Rented: Holiday vs. Working Day', fontsize=16)
plt.show()
```

**Insight:**
- Hanya selisih 2.8% antara total sewa sepeda pada working day dan holiday
